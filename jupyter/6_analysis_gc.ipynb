## Imports

```python
import pandas as pd

import sys 
sys.path.append('../python/')
import ap
import cfg
import db
```

## Data load and measure computation

```python
# "big table" with speaker normalized basic data and speaker pairs table
db.connect(cfg.CORPUS_ID_GC)
df_bt = ap.load_data(cfg.CORPUS_ID_GC, cfg.NRM_SPK)
df_spk_pairs = db.pd_read_sql_query(sql_fname=cfg.SQL_SP_FNAME_GC)
db.close()
```

```python
# local similarity
df_lsim = ap.lsim(df_bt)
```

```python
# synchrony
df_syn = ap.syn(df_bt)
```

```python
# local convergence
df_lcon = ap.lcon(df_bt)
```

```python
# global convergence
df_gcon, df_gcon_raw = ap.gcon(df_bt)
```

```python
# global similarity
df_gsim, df_gsim_raw = ap.gsim(df_bt, df_spk_pairs)
```
